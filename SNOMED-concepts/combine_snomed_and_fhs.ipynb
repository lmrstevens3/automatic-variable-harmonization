```python
import pandas as pd
```

```python
snomed = pd.read_csv("output/CONCEPT_with_classes.csv")
snomed
```

```python
snomed[snomed["concept_id"] == 4073050]
```

```python
# Transform to dbGaP
snomed = snomed.rename(columns={
    "concept_id": "variable_id",
    "concept_name": "variable_description",
    "concept_class_id": "dbGaP_dataset_id"
                         })
snomed["dbGaP_study_id"] = "SNOMED"
snomed = snomed[["variable_id", "variable_description", "dbGaP_dataset_id", "dbGaP_study_id"]]
snomed = snomed.astype({"variable_id": 'str'})
snomed
```

```python
fhs = pd.read_csv("../FHS_CHS_ARIC_MESA_dbGaP_var_doc_NLP.csv")
fhs
```

```python
combined_fhs_snomed = pd.concat([fhs, snomed])
study_dataset_var_id = "dbGaP_studyID_datasetID_varID"

combined_fhs_snomed[study_dataset_var_id] = combined_fhs_snomed["dbGaP_study_id"] + "." + combined_fhs_snomed["dbGaP_dataset_id"] + "." + combined_fhs_snomed["variable_id"]
combined_fhs_snomed
```

```python
combined_fhs_snomed.to_csv("output/combined_FHS_SNOMED.csv")
```
